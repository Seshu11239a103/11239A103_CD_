a) Bakery Billing System

```python
# Three Address Code
tac = [
    "t1 = price * qty",
    "t2 = t1 + tax"
]

print("Target Code:")
for line in tac:
    parts = line.split(" = ")
    left = parts[0]
    right = parts[1]
    print("MOV", left, ",", right)
```
